```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

---
# Column space of matrix
---

```python

# Matrix S: A 3x2 matrix representing two vectors in 3D space
S = np.array( [ [3, 0],
                [5, 2],
                [1, 2] ] )

# Vector v: A 3-dimensional vector
v = np.array([-3, 1, 5])
# You can try another vector by uncommenting the line below
# v = np.array([1, 7, 3])

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Step 1: Draw the plane corresponding to the column space of matrix S
# Create a grid of points (xx, yy) that span a range in the x and y directions
xx, yy = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))

# Calculate the normal vector to the plane using the cross product of the two columns of S
cp = np.cross(S[:,0], S[:,1])

# Calculate the corresponding z values for the plane using the plane equation: ax + by + cz = 0
z1 = (-cp[0]*xx - cp[1]*yy) / cp[2]

# Plot the plane as a semi-transparent surface in the 3D space
ax.plot_surface(xx, yy, z1, alpha=0.2)

# Step 2: Plot the two vectors from matrix S as lines starting from the origin
# First vector (corresponding to the first column of S)
ax.plot([0, S[0,0]], [0, S[1,0]], [0, S[2,0]], 'k')

# Second vector (corresponding to the second column of S)
ax.plot([0, S[0,1]], [0, S[1,1]], [0, S[2,1]], 'k')

# Step 3: Plot the vector v as a red line starting from the origin
ax.plot([0, v[0]], [0, v[1]], [0, v[2]], 'r')

# Adjust the view angle for better visualization
ax.view_init(elev=150, azim=0)

# Show the plot
plt.show()
```

# Row space of matrix
---

```python

# Matrix R: A 2x3 matrix representing two vectors in 3D space (row space)
R = np.array( [ [3, 5, 1],
                [0, 2, 2] ] )

# Vector v: A 3-dimensional vector
# v = np.array([-3, 1, 5])
# You can try another vector by uncommenting the line below
v = np.array([1, 7, 3])

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Step 1: Draw the plane corresponding to the row space of matrix R
# Create a grid of points (xx, yy) that span a range in the x and y directions
xx, yy = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))

# Calculate the normal vector to the plane using the cross product of the two rows of R
cp = np.cross(R[0,:], R[1,:])

# Calculate the corresponding z values for the plane using the plane equation: ax + by + cz = 0
z1 = (-cp[0]*xx - cp[1]*yy) / cp[2]

# Plot the plane as a semi-transparent surface in the 3D space
ax.plot_surface(xx, yy, z1, alpha=0.2)

# Step 2: Plot the two vectors from the rows of matrix R as lines starting from the origin
# First vector (corresponding to the first row of R)
ax.plot([0, R[0,0]], [0, R[0,1]], [0, R[0,2]], 'k')

# Second vector (corresponding to the second row of R)
ax.plot([0, R[1,0]], [0, R[1,1]], [0, R[1,2]], 'k')

# Step 3: Plot the vector v as a red line starting from the origin
ax.plot([0, v[0]], [0, v[1]], [0, v[2]], 'r')

# Adjust the view angle for better visualization
ax.view_init(elev=10, azim=0)

# Show the plot
plt.show()
```
