```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

---
# Making a matrix full-rank by "shifting"
---

```python
# Size of the matrix
m = 30

# Create the square symmetric matrix
A = np.random.randn(m, m)
A = np.round(10 * A.T @ A)

# Reduce the rank
A[:, 0] = A[:, 1]

# Shift amount (λ = lambda)
l = 0.01

# New matrix
B = A + l * np.eye(m)

# Print information
print('Rank (w/o shift) = %d' % np.linalg.matrix_rank(A))
print('Rank (with shift) = %d' % np.linalg.matrix_rank(B))

# Visualization
fig, ax = plt.subplots(1, 2, figsize=(14, 7))

# Heatmap of matrix A (without shift)
sns.heatmap(A, cmap='coolwarm', cbar=True, ax=ax[0], annot=False)
ax[0].set_title('Matrix A (w/o shift)')

# Heatmap of matrix B (with shift)
sns.heatmap(B, cmap='coolwarm', cbar=True, ax=ax[1], annot=False)
ax[1].set_title('Matrix B (with shift)')

plt.show()
```
