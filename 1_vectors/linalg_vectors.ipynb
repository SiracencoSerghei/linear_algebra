# Vector-vector multiplication: the dot product

```python
## many ways to compute the dot product

# v1 = np.array([ 1, 2, 3, 4, 5, 6 ])
v1 = np.array([ 1, 2, 3, 4, 5 ])
v2 = np.array([ 0, -4,  -3, 6, 5 ])

# method 1
dp1 = sum( np.multiply(v1,v2) )

# method 2
dp2 = np.dot( v1,v2 )

# method 3
dp3 = np.matmul( v1,v2 )

# method 4
dp4 = 0  # initialize

# loop over elements
for i in range(len(v1)):

    # multiply corresponding element and sum
    dp4 = dp4 + v1[i]*v2[i]


print(dp1,dp2,dp3,dp4)
```

# Dot product properties: associative and distributive

```python
## Distributive property

# create random vectors
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot( a , (b+c) )
res2 = np.dot(a,b) + np.dot(a,c)

# compare them
print([ res1,res2 ])
```

```python
## Associative property

# create random vectors
n = 5
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot( a , np.dot(b,c) )
res2 = np.dot( np.dot(a,b) , c )

# compare them
print(res1)
print(res2)


### special cases where associative property works!
# 1) one vector is the zeros vector
# 2) a==b==c
```

# Is the Dot product commutative?

```python
A = np.random.randn(100)
B = np.random.randn(100)
# print(A)
# print(' ')
# print(B)

a = np.dot(A, B)
b = np.dot(B, A)
print(a, ' ', b, '\n ', a - b)

A = np.array([2, 4])
B = np.array([3, 5])

print(A)
print(B, '\n ')

v = np.dot(A, B.T)
w = np.dot(B, A.T)
print(v, ' ', w, '\n', v-w)
```

# Vector length

```python
## a vector
v1 = np.array([ 1, 2, 3, 4, 5, 6 ])

## methods 1-4, just like with the regular dot product, e.g.:
vl1 = np.sqrt( sum( np.multiply(v1,v1)) )

## method 5: take the norm
vl2 = np.linalg.norm(v1)

print(vl1,vl2, '\n', vl2-vl1)
```

# The dot product from a geometric perspective

```python
## two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])

## compute the angle (radians) between two vectors
ang = np.arccos( np.dot(v1,v2) / (np.linalg.norm(v1)*np.linalg.norm(v2)) )


## draw them
fig = plt.figure()
ax = plt.axes(projection='3d')

# Plot the vectors
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b', label='v1')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'r', label='v2')

# Set the limits for each axis
ax.set_xlim([-6, 6])
ax.set_ylim([-6, 6])
ax.set_zlim([-6, 6])

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title(f'Angle between vectors: {ang:.2f} rad.')

# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
## equivalence of algebraic and geometric dot product formulas

### two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])


### algebraic
dp_a = np.dot( v1,v2 )

### geometric
dp_g = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)

### print dot product to command
print(dp_a)
print(dp_g)
```

# Cauchy-Schwartz inequality

```python
## create three vectors, one of which is dependent on another
a = np.random.randn(5)
b = np.random.randn(5)
c = np.random.randn(1)*a
print(a)
print(c)

# compute their dot product
aTb = np.dot(a,b)
aTc = np.dot(a,c)

# demonstrate the (in)equalities
print(f'{np.abs(aTb):.4f}, {np.linalg.norm(a)*np.linalg.norm(b):.4f}')
print(f'{np.abs(aTc):.4f}, {np.linalg.norm(a)*np.linalg.norm(c):.4f}')
```

# Vector outer product

```python
v1 = np.array([  1, 2, 3 ])
v2 = np.array([ -1, 0, 1 ])

# outer product
o_p = np.outer(v1,v2)

# terrible programming, but helps conceptually:
op = np.zeros((len(v1),len(v2)))
for i in range(0,len(v1)):
    for j in range(0,len(v2)):
        op[i,j] = v1[i] * v2[j]

print(o_p)
print(op)
```

# Vector cross product

```python
## create vectors
v1  = [ -3,  2, 5 ]
v2  = [  4, -3, 0 ]

## Python's cross-product function
v3a = np.cross( v1,v2 )

## "manual" method
v3b = [ [v1[1]*v2[2] - v1[2]*v2[1]],
        [v1[2]*v2[0] - v1[0]*v2[2]],
        [v1[0]*v2[1] - v1[1]*v2[0]] ]

print(v3a,v3b)


fig = plt.figure()
## ax = fig.add_subplot(projection='3d')
ax = plt.axes(projection = '3d')

## draw plane defined by span of v1 and v2
xx, yy = np.meshgrid(np.linspace(-10,10,10),np.linspace(-10,10,10))
z1 = (-v3a[0]*xx - v3a[1]*yy)/v3a[2]
ax.plot_surface(xx,yy,z1,alpha=.2)

## plot the two vectors
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'k')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'green')
ax.plot([0, v3a[0]],[0, v3a[1]],[0, v3a[2]],'r')


ax.view_init(azim=150,elev=45)
plt.show()
```

# Hermitian transpose (a.k.a. conjugate transpose)

```python
## create a complex number
z = complex(3,4)
print(z)
## magnitude (length)
print( np.linalg.norm(z) )

## by transpose?
print( np.transpose(z)*z )

## by Hermitian transpose
print( np.transpose(z.conjugate())*z )


## complex vector
v = np.array( [ 3, 4j, 5+2j, complex(2,-5) ] )
print(v)
print( v.T )
print( np.transpose(v) )
print( np.transpose(v.conjugate()) )
```

# Dot product with Unit vectors

```python
## create two random-integers vectors (R4)
n = 4
v1 = np.round(20*np.random.randn(4))
v2 = np.round(20*np.random.randn(4))
print(v1, '\n', v2)

## compute lengths of the individual vectors,
## and the magnitude of their dot product

v1m = np.sqrt(np.dot(v1, v1))
v2m = np.sqrt(np.dot(v2, v2))
dpm = np.abs(np.dot(v1, v2))
print(v1m, v2m, dpm)
print(' ')

## "normalize" the vectors
v1u = v1/v1m
v2u = v2/v2m
print(v1u,'\n', v2u)
print(' ')

## compute magnitude of that dot product
dpm = np.abs(np.dot(v1u, v2u))

print(dpm)
```
